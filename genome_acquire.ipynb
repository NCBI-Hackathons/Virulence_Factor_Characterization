# Acquiring genomes containing known virulence factors

This script will:
A) Extract the species names of known, experimentally verified virulence factors from the VFDB core database and put them into a dictionary
B) Take the species name dictionary, perform a search of the NCBI microbial genome assemblies
C) Download and package all those genomes (hopefully)

## Setup
import libraries, etc.

```python
from Bio import SeqIO
```

## FASTA format parser
https://bioperl.org/formats/sequence_formats/FASTA_sequence_format
top line looks like: 
">VFG037386(gb|YP_001847242) (bauE) ferric siderophore ABC transporter, ATP-binding protein BauE [Acinetobactin (VF0467)] [Acinetobacter baumannii ACICU]"

Since it's in a weird FASTA format that BioPython doesn't support, I'll have to parse the description information for the bug name, then ask genbank for the genome info in the assembly db


```python
gbid=[]
bugid=[]
for seq_record in SeqIO.parse("VFDB_setA_nt.fas", "fasta"):
    #print(seq_record.id[13:-1])
    #print(repr(seq_record.seq))
    #print(len(seq_record))
    x= seq_record.description.rsplit("[",1);
    y=x[-1]
    y=y[:-1]
    print(y)
    gbid.append(seq_record.id[13:-1])
    bugid.append(y)
    
    
    
    
```
